## Texas

```python
df_tx_get_columns = pd.read_excel('data/county/texas.xlsx', inferSchema=True, skiprows=2)
```

```python
columns = []

for column in df_tx_get_columns.columns:
    columns.append(column.replace('Cases', ' ').lstrip())
```

```python
df_tx = pd.read_excel('data/county/texas.xlsx', inferSchema=True, names=columns, skiprows=2)
```

```python
print('Number of Rows: {}'.format(df_tx.shape[0]))
print('Number of Columns: {}'.format(df_tx.shape[1]))
```

```python
df_tx.head()
```

```python
df_tx.tail()
```

```python
drop_rows = df_tx.iloc[254:]
```

```python
df_tx = df_tx.drop(drop_rows.index, axis=0)
```

```python
df_tx.tail()
```

```python
df_tx.dtypes
```

```python
df_tx.isnull().sum()
```
